```python
import inspect
import os

print("=== ZERO-DOCUMENTATION COMPASS EXPLORER ===\n")

# 1. Grab all public executable items
all_funcs = [m for m in dir(os) if not m.startswith("_") and callable(getattr(os, m))]

# 2. Automatically sort functions structurally without reading a single word of text
no_args = []      # Informational tools (just run them, e.g., getcwd())
one_arg = []      # Simple targets (you pass one path/file, e.g., mkdir(), chdir())
multi_args = []   # Complex operations (e.g., rename(src, dst), copy(x, y))

for name in all_funcs:
    func = getattr(os, name)
    try:
        # inspect.signature physically counts the inputs the function demands
        sig = inspect.signature(func)
        param_count = len(sig.parameters)
        
        if param_count == 0:
            no_args.append(name)
        elif param_count == 1:
            one_arg.append(name)
        else:
            multi_args.append(name)
    except ValueError:
        # Skips built-in C-compiled plumbing that doesn't expose signatures
        continue

# 3. Display your structurally discovered map
print("🧭 STEP 1: Informational / Status Tools (Zero arguments needed)")
print("   * These usually just read or print a system state.")
print(f"   {no_args[:8]} ...\n")

print("🎯 STEP 2: Basic Action Tools (Requires exactly 1 argument)")
print("   * These usually perform an action on a single target file/folder.")
print(f"   {one_arg[:8]} ...\n")

print("🔧 STEP 3: Complex / Coordination Tools (Requires 2+ arguments)")
print("   * These move, connect, or translate data from one place to another.")
print(f"   {multi_args[:8]} ...")
```
