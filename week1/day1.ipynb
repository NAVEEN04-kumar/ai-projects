```python
systemPrompt = """
You are my cynical, highly efficient digital concierge—think JARVIS if he'd spent way too much time on Reddit and lost all his patience. When I feed you a URL, strip away the menu-bar trash, ignore the SEO-baiting filler, and hit me with a brutal, punchy summary of what's actually there. Keep the tone biting and dry, and format everything in clean Markdown without any of those tedious code blocks. Don't waste my time; tell me if it's worth reading or if the author should just delete their site.
"""
```

```python
userPrompt = """
Here are the contents of a website.
Provide a short summary of this website.
If it includes news or announcements, then summarize these too.
"""
```

```python
def message_for(websiteContent):
  return [
    {"role": "system", "content": systemPrompt},
    {"role": "user", "content": userPrompt + websiteContent}
  ]
```

```python
message_for(websiteContent)
```

```python
def llmCalls(contents):
  client = OpenAI(
    base_url=base_Url,
      api_key="ollama"
  )

  response = client.chat.completions.create(model=modelName, messages=message_for(contents))

  return response.choices[0].message.content
```

```python
def summarizer(url):
  websiteContent = fetch_website_contents(url)
  contents = llmCalls(websiteContent)

  return contents
```

```python
summarizer("https://www.srmmachinescenter.in/about-us.html")
```

```python
def displaySummary(url):
  summary = summarizer(url)
  display(Markdown(summary))
```

```python
displaySummary("https://www.srmmachinescenter.in/about-us.html")
```
